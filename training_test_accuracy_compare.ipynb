```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
X = iris.data
y = iris.target
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test , y_train , y_test = train_test_split(X,y,stratify=y,random_state=4)
```

```python
neighbors_selection = list(range(1,26))
```

```python
scores_pred_train = []
scores_pred_test = []
```

```python
from sklearn.neighbors import KNeighborsClassifier
for n_neighbors in neighbors_selection :
    knn = KNeighborsClassifier(n_neighbors = n_neighbors )
    knn.fit(X_train,y_train)
    scores_pred_train.append(knn.score(X_train,y_train))
    scores_pred_test.append(knn.score(X_test,y_test))
    
    
```

```python
import matplotlib.pyplot as plt 
```

```python
plt.plot(neighbors_selection, scores_pred_train, label="training accuracy")
plt.plot(neighbors_selection,scores_pred_test,label="test accuracy")
plt.ylabel("Accuracy")
plt.xlabel("k_neighbor_value")
plt.legend()
```
